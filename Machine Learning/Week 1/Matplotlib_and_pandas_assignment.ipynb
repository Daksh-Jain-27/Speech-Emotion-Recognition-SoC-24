```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("company_sales_data.csv")
totprof = df['total_profit']
month = df['month_number']
plt.plot(month,totprof,'r:',label = "Total Profit",marker = 'o',lw = 3)
plt.legend(loc='lower right')
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
```

```python
plt.plot(month,df['facecream'],c="r")
plt.plot(month,df['facewash'],c="green")
plt.plot(month,df['toothpaste'],c="yellow")
plt.plot(month,df['bathingsoap'],c="m")
plt.plot(month,df['shampoo'],c="b")
# plt.plot(month,df['moisturizer'],c="black")
```
